$ \textbf{Question 1}$: If we list all the natural numbers below 10 that are multiples of 3 or 5, we get 3, 5, 6 and 9. The sum of these multiples is 23.

Find the sum of all the multiples of 3 or 5 below 1000.

```python
count =0
for i in range(1,1000):
    
    if i%3==0 or i%5==0:
        count+=i
print(count)
```

$ \textbf{Question 4}$: A palindromic number reads the same both ways. The largest palindrome made from the product of two 2-digit numbers is 9009 = 91 × 99.

Find the largest palindrome made from the product of two 3-digit numbers.


```python
def check_palindrome(num):
    num=str(num)
    return num==num[::-1]
list_of_palindromes=[] #of products of 3 digit no.s only
for x in range(100,1000):
    for y in range(100,1000):
        product=x*y
        if check_palindrome(product):
            list_of_palindromes.append(product)
        else:
            continue
            
max(list_of_palindromes)
```

Hence the difference between the sum of the squares of the first ten natural numbers and the square of the sum is\

$\textbf{Question 9}$ : A Pythagorean triplet is a set of three natural numbers, a < b < c, for which,

$a^2 + b^2 = c^2$ \
For example, \
$3^2 + 4^2 = 9 + 16 = 25 = 52.$

There exists exactly one Pythagorean triplet for which $a + b + c = 1000.$
Find the product abc

```python
def check_pyth(a,b,c):
    return (a**2+b**2)==c**2
lst=[]
for a in range(1,1000):
    for b in range(a,1000):
        c=1000-(a+b)
        if check_pyth(a,b,c):
            lst=[a,b,c]
a=lst[0]
b=lst[1]
c=lst[2]
print(a*b*c)
print(lst)
            
```

$\textbf{Question 10}$ : The sum of the primes below 10 is 2 + 3 + 5 + 7 = 17.

Find the sum of all the primes below two million.




```python
lst_of_primes=[2]
def check_prime(num):
    for prime in lst_of_primes:
        if num%prime==0:
            return False
    #print(num) #this is just to get a list of prime numbers for other purposes
    lst_of_primes.append(num)
    return True
sum_of_primes=2
n=2*(10**6)
for i in range(3,n,2):
    if check_prime(i):
        sum_of_primes+=i
print(sum_of_primes)
    
    
```
